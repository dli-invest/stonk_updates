```python
import pandas as pd

exchange = "NYE"
stock_df = pd.read_csv("https://raw.githubusercontent.com/dli-invest/eod_tickers/main/data/us_stock_data.csv")
print(stock_df)
# use whylogs to analysis csv
```

```python
!pip install pybars3
```

```python
import whylogs as why
from whylogs.core.constraints import Constraints, ConstraintsBuilder, MetricsSelector, MetricConstraint

print(stock_df.columns)
results = why.log(stock_df)

prof_view = results.view()

builder = ConstraintsBuilder(prof_view)

builder.add_constraint(MetricConstraint(
    name="price > 0",
    condition=lambda x: x.min >= 0,
    metric_selector=MetricsSelector(metric_name='distribution', column_name='price')
))
constraints: Constraints = builder.build()
constraints_report = constraints.report()
print(f"Constraints report [constraint name, pass, fail]: {constraints_report}")

from whylogs.viz import NotebookProfileVisualizer

# You can also pass the constraints to the NotebookProfileVisualizer and generate a report
visualization = NotebookProfileVisualizer()
visualization.constraints_report(constraints, cell_height=300)
```
